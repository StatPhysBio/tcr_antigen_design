```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

from scipy.stats import spearmanr

import sys
sys.path.append('../../src/')
from substitution_matrices import BLOSUM62

def robust_spearmanr(df_matrix1, df_matrix2, return_points=False):
    matrix1_flat = df_matrix1.values.flatten()
    matrix2_flat = df_matrix2.values.flatten()

    # they differ sometimes because two properties have a NaN
    mask1 = ~np.isnan(matrix1_flat)
    mask2 = ~np.isnan(matrix2_flat)
    mask = np.logical_and(mask1, mask2)

    matrix1_flat = matrix1_flat[mask]
    matrix2_flat = matrix2_flat[mask]

    return_value = spearmanr(matrix1_flat, matrix2_flat)
    
    if return_points:
        return_value = (return_value, (matrix1_flat, matrix2_flat))

    return return_value

AMINOACIDS = 'M F I L V W Y A G T S N Q D E H R K P'.split() # excluding C because Andreas doesn't consider it

df_blosum62 = pd.DataFrame(np.full((19, 19), np.nan), index=AMINOACIDS, columns=AMINOACIDS)
for aa1 in AMINOACIDS:
    for aa2 in AMINOACIDS:
        if aa1 == aa2:
            df_blosum62.loc[aa1, aa2] = np.nan
        else:
            df_blosum62.loc[aa1, aa2] = BLOSUM62[aa1][aa2] # negative so correlations are positive?

df_hermes_000 = pd.read_csv('../auxiliary_data/hcnn_fixed_structure/hcnn_pyrosetta_noise=0p0_marginal_matrix.csv', index_col=0)
df_hermes_000 = df_hermes_000.loc[AMINOACIDS, AMINOACIDS]
for aa1 in AMINOACIDS:
    for aa2 in AMINOACIDS:
        if aa1 == aa2:
            df_hermes_000.loc[aa1, aa2] = np.nan
        else:
            df_hermes_000.loc[aa1, aa2] = df_hermes_000.loc[aa1, aa2]

df_hermes_000_generic = pd.read_csv('../auxiliary_data/hcnn_fixed_structure/hermes_000_on_general_proteins_marginal_matrix.csv', index_col=0)
df_hermes_000_generic = df_hermes_000_generic.loc[AMINOACIDS, AMINOACIDS]
for aa1 in AMINOACIDS:
    for aa2 in AMINOACIDS:
        if aa1 == aa2:
            df_hermes_000_generic.loc[aa1, aa2] = np.nan
        else:
            df_hermes_000_generic.loc[aa1, aa2] = df_hermes_000_generic.loc[aa1, aa2]

df_AW_d = pd.read_csv('../auxiliary_data/AW_d.csv', index_col=0, sep='\t')
df_AW_d = -df_AW_d.loc[AMINOACIDS, AMINOACIDS]

hydrophobic_props = [f'V{i}' for i in range(1, 19)]
steric_props = [f'V{i}' for i in range(19, 36)]
electronic_props = [f'V{i}' for i in range(36, 51)]
props_type_dict = {
    'electronic': electronic_props,
    'hydrophobic': hydrophobic_props,
    'steric': steric_props,
}
prop_types = list(props_type_dict.keys())
all_props = hydrophobic_props + steric_props + electronic_props

prop_type_colors = {
    'electronic': 'blue',
    'hydrophobic': 'black',
    'steric': 'red',
}

df_aa_prop = pd.read_csv('../auxiliary_data/aminoacid_properties.csv')
df_aa_prop = df_aa_prop.drop(columns={'AA'})
df_aa_prop = df_aa_prop.set_index('abbr')

# make a pseudo-substitution matrix for each property
props_dfs = {}
for prop in df_aa_prop.columns:
    properties = dict(zip(df_aa_prop.index.values, df_aa_prop[prop].values))
    matrix = np.full((19, 19), np.nan)
    for i, aa1 in enumerate(AMINOACIDS):
        for j, aa2 in enumerate(AMINOACIDS):
            if aa1 == aa2:
                matrix[i, j] = np.nan
            else:
                matrix[i, j] = -np.abs(properties[aa1] - properties[aa2])
    props_dfs[prop] = pd.DataFrame(matrix, index=AMINOACIDS, columns=AMINOACIDS)

model_to_dfs = {
    'hermes': df_hermes_000,
    'hermes_generic': df_hermes_000_generic,
    'AW_d': df_AW_d,
    'blosum62': df_blosum62
}

correlations = {}
correlations['hermes'] = {prop: robust_spearmanr(df_hermes_000, props_dfs[prop]) for prop in all_props}
correlations['hermes_generic'] = {prop: robust_spearmanr(df_hermes_000_generic, props_dfs[prop]) for prop in all_props}
correlations['AW_d'] = {prop: robust_spearmanr(df_AW_d, props_dfs[prop]) for prop in all_props}
correlations['blosum62'] = {prop: robust_spearmanr(df_blosum62, props_dfs[prop]) for prop in all_props}
all_models = list(correlations.keys())

model_to_pretty_name_sr = {
    'hermes': r'$\rho_{H_f}$',
    'hermes_generic': r'$\rho_{H_{gen}}$',
    'blosum62': r'$\rho_{B_{62}}$',
    'AW_d': r'$-\rho_{\mathcal{A}}$'
}

model_to_pretty_name = {
    'hermes': r'$H_f$',
    'hermes_generic': r'$H_{gen}$',
    'blosum62': r'$B_{62}$',
    'AW_d': r'$-\mathcal{A}$'
}

model_pairs = [('hermes', 'hermes_generic'), ('hermes', 'AW_d'), ('hermes', 'blosum62'), ('AW_d', 'blosum62')]

fontsize = 18

ncols = len(model_pairs)
nrows = 1
colsize = 3.7
rowsize = 3.5
fig, axs = plt.subplots(figsize=(ncols*colsize, nrows*rowsize), ncols=ncols, nrows=nrows)

for col, (m1, m2) in enumerate(model_pairs):

    ax = axs[col]

    for prop_type in prop_types:
        m1_values = [correlations[m1][prop][0] for prop in props_type_dict[prop_type]]
        m2_values = [correlations[m2][prop][0] for prop in props_type_dict[prop_type]]
        ax.scatter(m1_values, m2_values, facecolor='none', edgecolor=prop_type_colors[prop_type], marker='o', label=prop_type)

        # plot centroid
        centroid_m1 = np.mean(m1_values)
        centroid_m2 = np.mean(m2_values)
        ax.scatter(centroid_m1, centroid_m2, color=prop_type_colors[prop_type], marker='X', s=100)
    
    ax.axvline(0.0, ls='--', color='dimgrey')
    ax.axhline(0.0, ls='--', color='dimgrey')
    
    ax.set_xlabel(model_to_pretty_name_sr[m1], fontsize=fontsize)
    ax.set_ylabel(model_to_pretty_name_sr[m2], fontsize=fontsize)

    ax.tick_params(axis='both', labelsize=fontsize-4)

# set xlim and ylim of every subplot to be the same
xlims = [ax.get_xlim() for ax in axs]
ylims = [ax.get_ylim() for ax in axs]

xlim = (min([xlimm[0] for xlimm in xlims]), max([xlimm[1] for xlimm in xlims]))
ylim = (min([ylimm[0] for ylimm in ylims]), max([ylimm[1] for ylimm in ylims]))

lim = (min(xlim[0], ylim[0]), max(xlim[1], ylim[1]))

for ax in axs:
    ax.set_xlim(lim)
    ax.set_ylim(lim)
    ax.set_xticks([-0.2, 0.0, 0.2, 0.4, 0.6])
    ax.set_yticks([-0.2, 0.0, 0.2, 0.4, 0.6])
    ax.plot(lim, lim, ls=':', color='dimgrey')


plt.tight_layout()
plt.savefig('../auxiliary_data/scatterplot_vs_properties.png')
plt.savefig('../auxiliary_data/scatterplot_vs_properties.pdf')
plt.show()

# plot legend on separate figure
legend_handles = ax.get_legend_handles_labels()[0]
fig_leg = plt.figure(figsize=(8, 2))
ax_leg = fig_leg.add_subplot(111)
ax_leg.legend(handles=legend_handles, fontsize=fontsize-3, loc='center')
ax_leg.axis('off')
plt.tight_layout()
plt.savefig('../auxiliary_data/scatterplot_vs_properties__legend.png')
plt.savefig('../auxiliary_data/scatterplot_vs_properties__legend.pdf')
plt.show()


plt.figure(figsize=(5, 5))
for i, prop_type in enumerate(prop_types):
    m_values = [correlations['hermes_generic'][prop][0] for prop in props_type_dict[prop_type]]
    parts = plt.violinplot([m_values], positions=[i], widths=0.6, showmedians=True, showextrema=False)
    parts['cmedians'].set_color(prop_type_colors[prop_type])
    parts['cmedians'].set_linewidth(2)
    for pb, color in zip(parts['bodies'], [prop_type_colors[prop_type]]):
        pb.set_facecolor(color)
        pb.set_edgecolor(color)
        pb.set_alpha(0.5)
plt.axhline(0, ls='--', color='black')
plt.grid(axis='y', ls='--', alpha=0.5)
plt.xticks(range(len(prop_types)), prop_types, rotation=70, fontsize=fontsize-3)
plt.yticks(fontsize=fontsize-3)
plt.title('Spearman r\nbetween HERMES 0.00 subst-matrix\nand property subst-matrix', fontsize=fontsize-2)
plt.ylabel('Spearman r', fontsize=fontsize-2)
plt.tight_layout()
plt.savefig('../auxiliary_data/violinplot_hermes_generic_vs_properties.png')
plt.savefig('../auxiliary_data/violinplot_hermes_generic_vs_properties.pdf')
plt.show()



# compute correlations between matrices now, plot scatterplots
ncols = len(model_pairs)
nrows = 1
colsize = 3.7
rowsize = 3.5
fig, axs = plt.subplots(figsize=(ncols*colsize, nrows*rowsize), ncols=ncols, nrows=nrows)

for col, (m1, m2) in enumerate(model_pairs):

    ax = axs[col]

    (sr, sr_pval), (m1_values, m2_values) = robust_spearmanr(model_to_dfs[m1], model_to_dfs[m2], return_points=True)

    ax.scatter(m1_values, m2_values, color='black', alpha=0.5)
    
    ax.set_xlabel(model_to_pretty_name[m1], fontsize=fontsize)
    ax.set_ylabel(model_to_pretty_name[m2], fontsize=fontsize)

    ax.tick_params(axis='both', labelsize=fontsize-4)

    # print spearmanr in bottom right corner
    ax.text(0.02, 0.98, rf'$\rho = {sr:.2f}$', transform=ax.transAxes, fontsize=fontsize-2, ha='left', va='top')

plt.tight_layout()
plt.savefig('../auxiliary_data/scatterplot_matrices_against_each_other.png')
plt.savefig('../auxiliary_data/scatterplot_matrices_against_each_other.pdf')
plt.show()



# plot the matrices!

model_to_pretty_name_2 = {
    'hermes': r'$H_f - \overline{H_f}$',
    'hermes_generic': r'$H_{gen} - \overline{H_{gen}}$',
    'blosum62': r'$B_{62} - \overline{B_{62}}$',
    'AW_d': r'$-\mathcal{A} - \overline{-\mathcal{A}}$'
}

fontsize = 18

for col, m in enumerate(all_models):

    matrix = model_to_dfs[m].values
    matrix -= np.nanmean(matrix)

    mmin = np.nanmin(matrix)
    mmax = np.nanmax(matrix)
    vmin = min(mmin, -mmax)
    vmax = max(mmax, -mmin)

    fig = plt.figure(figsize=(6, 6))
    data = plt.imshow(matrix, cmap='coolwarm', vmin=vmin, vmax=vmax)
    plt.title(model_to_pretty_name_2[m], fontsize=fontsize+2)
    plt.xticks(np.arange(19), list(AMINOACIDS), fontsize=fontsize-2)
    plt.yticks(np.arange(19), list(AMINOACIDS), fontsize=fontsize-2)
    colorbar = fig.colorbar(data)
    colorbar.ax.tick_params(labelsize=fontsize+5)
    # colorbar.set_label(model_to_pretty_name_2[m], fontsize=fontsize+10)
    plt.tight_layout()
    plt.savefig(f'../auxiliary_data/average_matrix_{m}.png')
    plt.savefig(f'../auxiliary_data/average_matrix_{m}.pdf')
    plt.show()


# ## plot the HERMES matrices with the diagonal and the cysteine as well

# AMINOACIDS = 'C M F I L V W Y A G T S N Q D E H R K P'.split()

# df_hermes_000_full = pd.read_csv('../auxiliary_data/hcnn_fixed_structure/hcnn_pyrosetta_noise=0p0_marginal_matrix.csv', index_col=0)
# df_hermes_000_full = df_hermes_000_full.loc[AMINOACIDS, AMINOACIDS]

# df_hermes_050_full = pd.read_csv('../auxiliary_data/hcnn_fixed_structure/hcnn_pyrosetta_noise=0p5_marginal_matrix.csv', index_col=0)
# df_hermes_050_full = df_hermes_050_full.loc[AMINOACIDS, AMINOACIDS]


# for name, matrix_df in zip(['hermes_fixed_000_full', 'hermes_fixed_050_full'], [df_hermes_000_full, df_hermes_050_full]):

#     matrix = matrix_df.values
#     # matrix -= np.nanmean(matrix)

#     # mmin = np.nanmin(matrix)
#     # mmax = np.nanmax(matrix)
#     # vmin = min(mmin, -mmax)
#     # vmax = max(mmax, -mmin)

#     # print(np.nanmax(matrix))

#     from matplotlib.colors import TwoSlopeNorm

#     norm = TwoSlopeNorm(vmin=np.nanmin(matrix), vcenter=0, vmax=np.nanmax(matrix))

#     fig = plt.figure(figsize=(6, 6))
#     data = plt.imshow(matrix, cmap='coolwarm', norm=norm, interpolation='none') #, vmin=vmin, vmax=vmax)
#     # plt.title(model_to_pretty_name[m], fontsize=fontsize)
#     plt.xticks(np.arange(20), list(AMINOACIDS), fontsize=fontsize-2)
#     plt.yticks(np.arange(20), list(AMINOACIDS), fontsize=fontsize-2)
#     colorbar = fig.colorbar(data)
#     colorbar.ax.tick_params(labelsize=fontsize+5)
#     colorbar.set_label(name, fontsize=fontsize+10)
#     plt.tight_layout()
#     plt.savefig(f'../auxiliary_data/average_matrix_{name}.png')
#     plt.savefig(f'../auxiliary_data/average_matrix_{name}.pdf')
#     plt.show()
```
